```

telone= pd.read_csv("/content/Telkom 1.csv")

telone.rename(columns={'PRODUTC': 'PRODUCT', 'DATETIME': 'DATE_TIME'}, inplace=True)
telone.head()
```

```
teltwo= pd.read_csv('/content/Telkom 2.csv')

teltwo.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT', 'DW_B_NUMBER':'DW_B_NUMBER_INT'}, inplace=True)
teltwo.head()
```

```
telthree= pd.read_csv("/content/Telkom 3.csv")

telthree.rename(columns={'CELLID': 'CELL_ID', 'SIET_ID':'SITE_ID'}, inplace=True)
telthree.tail()
# telthree.info()
```

```

df=pd.concat([telone, teltwo, telthree], ignore_index=True)
df.tail()
# df.info()

# df.drop(['COUNTRY_A','COUNTRY_B','CELL_ON_SITE','DW_A_NUMBER_INT', 'DW_B_NUMBER_INT'], axis=1, inplace= True)
```

```
df.isnull().sum()
```

```
df.drop(['CELL_ON_SITE', 'DW_A_NUMBER_INT','DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B'], axis=1, inplace=True)
df.head()
```

```
df = df.groupby(['PRODUCT'], as_index=False)['VALUE'].sum()
df
```

```
df = df.groupby(['PRODUCT'], as_index=False)['VALUE'].mean()
df
```

```
df = df.PRODUCT.value_counts()
df
```

```
df[['DATE', 'TIME']] = df.DATE_TIME.str.split(expand=True)
df.tail()

```
